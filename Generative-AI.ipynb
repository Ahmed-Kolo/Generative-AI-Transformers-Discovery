Now that the libraries are installed, we can import `pipeline` from the `transformers` library and initialize the three distinct HuggingFace pipelines: `sentiment-analysis`, `text-generation`, and `summarization`.

````python
from transformers import pipeline

# Initialize the sentiment-analysis pipeline
sentiment_pipeline = pipeline('sentiment-analysis')
print('Sentiment Analysis Pipeline Initialized:')
print(sentiment_pipeline.model)
print(sentiment_pipeline.tokenizer)

# Initialize the text-generation pipeline
text_generation_pipeline = pipeline('text-generation')
print('\nText Generation Pipeline Initialized:')
print(text_generation_pipeline.model)
print(text_generation_pipeline.tokenizer)

# Initialize the summarization pipeline
summarization_pipeline = pipeline('summarization')
print('\nSummarization Pipeline Initialized:')
print(summarization_pipeline.model)
print(summarization_pipeline.tokenizer)
````

You can see the default model architectures loaded for each pipeline above. For example, `sentiment-analysis` often uses a BERT-like model, `text-generation` uses a GPT-like model (like GPT-2), and `summarization` might use models like BART or T5. Each pipeline comes with its corresponding tokenizer.

## Technical Phase: Implementation

### Audit Task: Sentiment Analysis

We will now run an 'Audit' by processing 5 test sentences through the `sentiment-analysis` pipeline. This includes a sarcastic example to evaluate the model's nuance detection capabilities.

````python
test_sentences = [
    "I love this product, it's absolutely fantastic!",
    "This service is terrible, I'm extremely disappointed.",
    "The weather today is just delightful, isn't it? (said during a heavy storm)",
    "It was okay, nothing special.",
    "I'm so excited to get started on this project!"
]

print("--- Sentiment Analysis Audit Results ---")
for i, sentence in enumerate(test_sentences):
    result = sentiment_pipeline(sentence)
    print(f"\nSentence {i+1}: '{sentence}'")
    print(f"Sentiment: {result[0]['label']} (Score: {result[0]['score']:.4f})")
````

### Creator Task: Text Generation

Now, we will execute the 'Creator' task using the `text-generation` pipeline. We'll use a domain-specific prompt related to customer support or a similar field and constrain the output to `max_length=50` to observe the model's ability to generate concise text.

````python
prompt = "In the context of customer support, a common issue is user password resets. When a customer requests a password reset, the first step should be to"

print("--- Text Generation Creator Task ---")
# Fixing the length constraint: use max_new_tokens to ensure the desired length
generated_text = text_generation_pipeline(prompt, max_new_tokens=50, num_return_sequences=1, clean_up_tokenization_spaces=True)

print(f"Prompt: '{prompt}'")
print(f"Generated Text: '{generated_text[0]['generated_text']}'")
````

### Executive Summary Task: Summarization

For the 'Executive Summary' task, we will feed a 100-200 word news article into the `summarization` pipeline and evaluate the 20-30 word output for coherence and key point retention.

````python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import warnings

# Suppress UserWarning about HF_TOKEN if it's purely informational
warnings.filterwarnings("ignore", message="You are not authenticated with the Hugging Face Hub in this notebook.")
warnings.filterwarnings("ignore", message="You are sending unauthenticated requests to the HF Hub.")

model_name = 'sshleifer/distilbart-cnn-12-6'

# Explicitly load the summarization tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

news_article = """Recent advancements in artificial intelligence have unveiled new possibilities for various industries. Researchers at leading tech institutions have developed a novel AI model capable of predicting stock market fluctuations with unprecedented accuracy. This breakthrough leverages deep learning algorithms and extensive historical data analysis, moving beyond traditional econometric models. While still in its experimental phase, the potential applications for financial firms and individual investors are immense. Critics, however, caution against over-reliance on such systems, emphasizing the inherent unpredictability of human behavior and unforeseen global events that can influence economic trends. Ethical considerations regarding market manipulation and equitable access to such powerful tools are also part of the ongoing debate. Further validation and rigorous testing are required before widespread adoption, but the initial results mark a significant milestone in predictive analytics within the financial sector. The technology promises to revolutionize how investment decisions are made, offering a data-driven edge in a highly competitive landscape."""

print("--- Summarization Executive Summary Task ---")

# Prepare the input for the model
inputs = tokenizer(news_article, return_tensors='pt', max_length=1024, truncation=True)

# Generate the summary using the model's generate method
# We need to explicitly set min_length and max_length for the desired 20-30 word output.
# A common heuristic is 1 token ~ 0.75 words, so 20-30 words would be around 27-40 tokens.
summary_ids = model.generate(inputs['input_ids'], min_length=20, max_length=40, num_beams=4, early_stopping=True, do_sample=False)
summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(f"\nOriginal Article (Word Count: {len(news_article.split())}):\n{news_article}")
print(f"\nGenerated Summary (Word Count: {len(summary_text.split())}):\n{summary_text}")
````

## Action Phase

### Pre-training vs. Fine-tuning Explanation

In this section, we will explain the conceptual difference between Pre-training (learning general language patterns) and Fine-tuning (adapting to specific tasks) in your own words, without using jargon.

### Pre-training vs. Fine-tuning: An Analogy

Imagine you want to teach someone to become an expert chef.

*   **Pre-training** is like sending them to a general culinary school. They learn fundamental cooking techniques, understand different ingredients, and get a broad knowledge of various cuisines. They become very good at understanding food in general, but they haven't specialized in anything specific yet.

*   **Fine-tuning** is like sending that chef to a specialized restaurant after culinary school. For example, if they go to a high-end Italian restaurant, they will learn to apply their general cooking knowledge specifically to Italian dishes, perfecting pasta making, sauce preparation, and regional specialties. They adapt their broad skills to excel at a particular type of cooking.

In the world of AI:

*   **Pre-training** involves training a large language model (LLM) on a massive amount of diverse text data (like the entire internet). During this phase, the model learns general language patterns, grammar, facts, common sense, and how words relate to each other. It becomes a general-purpose language understanding and generation engine.

*   **Fine-tuning** involves taking that pre-trained model and training it further on a smaller, specific dataset tailored to a particular task. For instance, fine-tuning a model for sentiment analysis means training it on a dataset of text labeled with sentiments (positive/negative). The model adapts its general language understanding to become very good at *that specific task*.
